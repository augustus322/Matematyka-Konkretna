```python
import matplotlib.pyplot as plt
import numpy as np

# 1. Funkcja sigmoidalna
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

# 2. Gradient (pochodna) funkcji sigmoidalnej
def sigmoid_gradient(x):
    s = sigmoid(x)
    return s * (1 - s)

# 3. Zakres warto≈õci x
x = np.linspace(-5, 5, 200)
y = sigmoid(x)
dy = sigmoid_gradient(x)

# 4. Wykres
plt.figure(figsize=(10, 6))
plt.plot(x, y, 'b-', linewidth=2, label='Sigmoid $\sigma(z)$')
plt.plot(x, dy, 'r--', linewidth=2, label='Gradient $\sigma\'(z)$')
plt.axhline(y=0, color='k', linestyle='-', linewidth=0.5)
plt.axhline(y=1, color='k', linestyle=':', linewidth=0.5)
plt.axvline(x=0, color='k', linestyle='-', linewidth=0.5)
plt.xlabel('z')
plt.ylabel('Warto≈õƒá')
plt.title('Funkcja sigmoidalna i jej gradient')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
